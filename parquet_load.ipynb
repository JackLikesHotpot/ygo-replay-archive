````python
import pandas as pd
import re

RE_MATCH_INDICATOR = re.compile(  r'\bround\s*\d+|\btop\s*\d+|\bfinals?\b|\bsemi[-\s]?finals?\b|\bvs\.?\b|\b\d+(?:st|nd|rd|th)\s+place\b', re.IGNORECASE) 
RE_EXCLUDE = re.compile(r'\bannouncement\b|\binterview\b|\bdiscussion\b|\bvlog\b|\bdeck\s*profile\b', re.IGNORECASE)
RE_VS_PATTERN = re.compile(r'\bvs\.?\b', re.IGNORECASE)
RE_PLACEMENT = re.compile(r'\b\d+(?:st|nd|rd|th)\s+place\b', re.IGNORECASE)

df = pd.read_parquet("pipeline/checkpoints/stage4_videos.parquet")

match_mask   = df["Title"].str.contains(RE_MATCH_INDICATOR, na=False)
exclude_mask = df["Title"].str.contains(RE_EXCLUDE, na=False)

matches = df[match_mask & ~exclude_mask]
dropped = df[~match_mask | exclude_mask]

print(f"Kept: {len(matches)}")
print(f"Dropped: {len(dropped)}")
````

````python
RE_LIVESTREAM_DAY = re.compile(r'\bday\s*[123]\b', re.IGNORECASE)

livestreams = dropped[dropped["Title"].str.contains(RE_LIVESTREAM_DAY, na=False)]
dropped_remaining = dropped[~dropped["Title"].str.contains(RE_LIVESTREAM_DAY, na=False)]

print(f"Livestreams: {len(livestreams)}")
print(f"Dropped (remaining): {len(dropped_remaining)}")
````

````python
from groq import Groq
import os
import json
import pandas as pd
from dotenv import load_dotenv
load_dotenv()

client = Groq(api_key=os.getenv("GROQ_API_KEY"))

sample = llm_input.sample(n=20, random_state=56)
sample_records = sample[["Video ID", "Title", "Video Description"]].to_dict("records")

prompt = """
You are a Yu-Gi-Oh! feature match classification system.

You will receive a JSON array of videos. For EVERY video in the input, return EXACTLY ONE object in the output array.

A feature match is a real recorded duel between two identifiable players.
Livestreams, vlogs, deck profiles, promos, interviews, and commentary-only videos are NOT feature matches.

IMPORTANT: Titles containing round/bracket stage indicators (e.g. "Round X", "Top X", "Quarterfinal", 
"Semi-Final", "Final") almost always indicate a feature match, even if player names are not present 
in the title or description. Player names being unavailable does NOT mean it's not a match — 
set player_1/player_2/deck_1/deck_2 to null in that case, but still mark is_match as true.

Return ONLY a valid JSON array, one object per input video, in the same order, using this schema:

{
  "video_id": string,
  "is_match": true or false,
  "tournament": string or null,
  "round": string or null,
  "player_1": string or null,
  "deck_1": string or null,
  "player_2": string or null,
  "deck_2": string or null
}

RULES:
- ALWAYS include "video_id" exactly as given in the input.
- If is_match is false, set all other fields to null.
- Round must be formatted exactly like: "Round 1", "Top 32", "Finals", "Semi-Final", "Quarterfinal"
- Tournament should be the event name (e.g. "YCS London 2024")
- Do NOT guess missing players or decks — use null if unknown

INPUT:
""" + json.dumps(sample_records)

response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": prompt}],
    temperature=0,
    max_tokens=8192
)

content = response.choices[0].message.content.strip()

# strip markdown code fences if present
if content.startswith("```"):
    content = content.split("```")[1]
    if content.startswith("json"):
        content = content[4:]
    content = content.strip()

parsed = json.loads(content)
results_df = pd.DataFrame(parsed)
results_df
````

````python
prompt_template = """
You are a Yu-Gi-Oh! feature match classification system.

You will receive a JSON array of videos. For EVERY video in the input, return EXACTLY ONE object in the output array.

A feature match is a real recorded duel between two identifiable players.
Livestreams, vlogs, deck profiles, promos, interviews, and commentary-only videos are NOT feature matches.

IMPORTANT: Titles containing round/bracket stage indicators (e.g. "Round X", "Top X", "Quarterfinal", 
"Semi-Final", "Final") almost always indicate a feature match, even if player names are not present 
in the title or description. Player names being unavailable does NOT mean it's not a match — 
set player_1/player_2/deck_1/deck_2 to null in that case, but still mark is_match as true.

Return ONLY a valid JSON array, one object per input video, in the same order, using this schema:

{
  "video_id": string,
  "is_match": true or false,
  "tournament": string or null,
  "round": string or null,
  "player_1": string or null,
  "deck_1": string or null,
  "player_2": string or null,
  "deck_2": string or null
}


RULES:
- ALWAYS include "video_id" exactly as given in the input.
- If is_match is false, set all other fields to null.
- Round must be formatted exactly like: "Round 1", "Top 32", "Finals", "Semi-Final", "Quarterfinal"
- Tournament should be the event name (e.g. "YCS London 2024")
- Do NOT guess missing players or decks — use null if unknown
- DO NOT write Python or JavaScript.
- DO NOT explain.
- DO NOT wrap in ```json.
- DO NOT write any text before or after the JSON.
- The first character of your response MUST be [
- The last character of your response MUST be ]

INPUT:
"""

def safe_parse_json(content: str):
    if content.startswith("```"):
        content = content.split("```")[1]
        if content.startswith("json"):
            content = content[4:]
        content = content.strip()

    try:
        return json.loads(content)
    except json.JSONDecodeError as e:
        print(f"JSON parse failed: {e}")
        print(f"Raw content (last 300 chars): {content[-300:]}")
        return None
````

````python
import time

all_results = []
chunk_size = 10
rows = llm_input.to_dict("records")
total_chunks = -(-len(rows) // chunk_size)

for i in range(0, len(rows), chunk_size):
    chunk_id = i // chunk_size + 1
    chunk = rows[i:i + chunk_size]

    print(f"Processing chunk {chunk_id}/{total_chunks}")

    chunk_prompt = prompt_template + json.dumps(chunk, ensure_ascii=False)

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": chunk_prompt}],
        temperature=0,
        max_tokens=1000,
        response_format={"type": "json_object"}
    )

    content = response.choices[0].message.content.strip()
    parsed = safe_parse_json(content)

    if parsed is not None:
        all_results.extend(parsed)
        print(f"  ✓ {len(parsed)} rows returned")
    else:
        print(f"  ✗ chunk {chunk_id} failed — skipping")

    if i + chunk_size < len(rows):  # don't sleep after the last chunk
        time.sleep(30)
        # break

results_df = pd.DataFrame(all_results)
print(f"\nTotal rows processed: {len(results_df)} / {len(rows)}")
````
